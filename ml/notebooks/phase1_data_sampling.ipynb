```python
import pandas as pd
from pathlib import Path

# Define file paths
RAW_DIR = Path("../data_raw")        # this assumes notebook is inside ml/notebooks/
DATA_FILE = RAW_DIR / "ai4i2020.csv"
SAMPLE_FILE = RAW_DIR / "ai4i2020_sample.csv"

# Load the real dataset
df = pd.read_csv(DATA_FILE)
print(f"Full dataset shape: {df.shape}")

# Take 10% sample for fast iteration
sample_df = df.sample(frac=0.1, random_state=42)
sample_df.to_csv(SAMPLE_FILE, index=False)

print(f"Sample dataset shape: {sample_df.shape}")
print(f"✅ Sample saved to: {SAMPLE_FILE}")
```
